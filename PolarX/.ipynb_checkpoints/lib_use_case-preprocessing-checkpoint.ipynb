### Removing stripe artefacts on images

### Algorithm:
1. Cut central part of all images
2. Average over rows of each image (along 'x' coordinate)
3. For each line subtract its median (allows to set background intensity to zero)
4. Look at the distribution of all intensity points (all lines) and realize that fluctuation amplitude is approximately normally distributed. Calculate its mean and use it as an estimator for the noise amplitude.
5. Convert line intensity profile to a step mask (using thresholding) which takes the value of noise intensity at values 'y' where the intensity drops and zero otherwise. Mask is calculated separately for each image.
6. Add this mask to each image (for every column)

```python
x, lines, noise = bl.remove_artefact(x_raw, thr1=-3, thr2=-2.5, left=27, right=35)
```

```python
h = lines.plot.hist(bins=50)
axvline(x = -noise)
```

### Peak at zero corresponds to the backgroung intensity which is not affected by the noise. The peak on the left corresponds to the noise amplitude. Here we assume normal distribution of both and separate them simply based on threshold (thr1) parameter. 

PS Can be done in a more presice way by creating a model with two overlapping gaussians and having more accurate estimators for $\mu_1, \mu_2$

### Stabilization of images (image registration using cross-correlation). First we choose a reference image to register to. Extract images corresponding to the (30, 40) sec time interval (assuming that images are stable within this period) and average them. 

```python
ref_image = x.sel(t=slice(30,40)).mean('t')
ref_image.plot()
```

### Estimate shifts for each image

```python
shifts = bl.register_xarray(x, ref_image, n_aver=200)
```

### Apply shifts. Yields stabilized DataArray

```python
bl.shift_xarray(x, shifts)
```

### Cropping an image. Using sel method of a DataArray (numbers in slice object correspond to the real physical units - $\mu m$)

```python
x_cut = x.sel(x=slice(26, 37), y=slice(1, 11))
x_cut.mean('t').plot()
```

### Block average in time. Images are grouped into chunks of N and averaged. Resulting images are combined back into a DataArray

```python
x_aver = bl.chunk_average_in_time(x_cut, 300)
x_aver[0].plot()
```
